```python
import math
import matplotlib.pyplot as plt

from skimage import io
from skimage import data
from skimage import transform
from skimage.transform import swirl, warp
```

https://www.pexels.com/photo/abstract-art-circle-clockwork-414579/

```python
gears = io.imread('./images/pexels-gears.jpg')  

plt.figure(figsize=(8, 8))
plt.imshow(gears, cmap='gray')
```

###  Perform a similarity transform on the image

Specifies rotation, translation and scaling of the image

```python
tform = transform.SimilarityTransform(scale=1.5, rotation=math.pi/4,
                                      translation=(gears.shape[0]/2, -100))

rotated = warp(gears, tform)
back_rotated = warp(rotated, tform.inverse)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(rotated)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(back_rotated)
```

### Swirl transformation

* rotation = Additional rotation applied to the image
* strength = The amount of swirling applied
* radius = The extent of the swirl (in pixels)
* mode (reflect) = reflect the values of 1st and last pixels outside the boundary of input

```python
# Change radius to 350 and re-rum
# Change strength to 10 and re-run
gears_swirled = swirl(gears, rotation=0, strength=20, radius=350, mode='constant')

plt.figure(figsize=(8, 8))
plt.imshow(gears_swirled)
```
